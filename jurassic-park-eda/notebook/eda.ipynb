```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

```python
dino = pd.read_csv('../data/processed/dinosaursclean.csv')
dino.head(1)
```

# What's the distribution of dinosaur diets 
- herbivorous
- carnivorous
- omnivorous

```python
sns.set(style='whitegrid')

diets = sns.countplot(dino, x='diet', order=dino['diet'].value_counts().index, palette='pastel')

diets.set_title("Distribution of Dinosaur Diets", fontsize=16, weight='bold')
diets.set_xlabel('Diet', fontsize=12)
diets.set_ylabel('Count', fontsize=12)

plt.xticks(rotation=30, ha='right')

for p in diets.patches:
    diets.annotate(f'{p.get_height()}', (p.get_x() + p.get_width() / 2., p.get_height()),
                ha='center', va='bottom', fontsize=10)
```

# Are certain diets more common in specific time periods?

```python
sns.set(style='whitegrid')

perioddiets = sns.countplot(data=dino, x='diet', hue='period name', order=dino['diet'].value_counts().index, palette='pastel')

perioddiets.set_title("Diets per Period", fontsize=16, weight='bold')
perioddiets.set_xlabel('Diet', fontsize=12)
perioddiets.set_ylabel('Count', fontsize=12)

plt.xticks(rotation=30, ha='right')
```

# Are some Orders or Classes more associated with certain diets?

```python
sns.set(style='whitegrid')

perioddiets = sns.countplot(data=dino, x='diet', hue='Class', order=dino['diet'].value_counts().index, palette='pastel')

perioddiets.set_title("Diets per Class", fontsize=16, weight='bold')
perioddiets.set_xlabel('Diet', fontsize=12)
perioddiets.set_ylabel('Count', fontsize=12)

plt.xticks(rotation=30, ha='right')
```

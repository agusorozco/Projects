```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

# Feature Engineering
Here I mess around with some of the features in the dataset.

```python
titanic = pd.read_csv('../data/titanic.csv')
titanic
```

```python
titanic['adult_male'] = titanic['adult_male'].map({True: 1, False: 0})
titanic['alone'] = titanic['alone'].map({True: 1, False: 0})
titanic
```

```python
titanic['child'] = (titanic['age'] <= 18).astype(int)
titanic
```

```python
titanic['sex'] = (titanic['sex'] == 'male').astype(int)
titanic
```

```python
titanic['deck'] = titanic['deck'].fillna('Unknown')

deck_dummies = pd.get_dummies(titanic['deck'], prefix='deck', drop_first=True)

titanic = pd.concat([titanic, deck_dummies], axis=1)
titanic.drop(columns='deck', inplace=True)
```

```python
titanic['embark_town'] = titanic['embark_town'].fillna('Unknown')

embark_dummies = pd.get_dummies(titanic['embark_town'], prefix='embark', drop_first=True)

titanic = pd.concat([titanic, embark_dummies], axis=1)
titanic.drop(columns='embark_town', inplace=True)
```

```python
titanic['family size'] = titanic['parch'] + titanic['sibsp']
```

```python
titanic.to_csv('cleanedtitanic.csv', index=False)
```
